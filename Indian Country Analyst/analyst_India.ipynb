## top_cities DataFrame'ini, 'population_total' sütununa göre büyükten küçüğe sıralayarak en yüksek toplam nüfusa sahip ilk 10 şehri seçer.

plt.figure(figsize=(12, 6)): 12x6 boyutlarında bir grafik figürü oluşturulur. Grafik figürü, içine eklenen çizimlerin boyutlarını belirler.

**Toplam nüfus bar grafiği:** plt.bar(top_cities['name_of_city'], top_cities['population_total'], color='green', label='Total Population'). Bu çubuk grafiği, seçilen şehirlerin isimlerine karşılık gelen toplam nüfus verilerini gösterir. Yeşil renk kullanılır ve grafiğin üzerinde 'Total Population' etiketi bulunur.

### Erkek nüfusu ve Kadın nüfusu bar grafikleri:

**plt.bar(top_cities['name_of_city'], top_cities['population_male'], color='blue', alpha=0.7, label='Male Population'):** Mavi renkte, erkek nüfusunu temsil eden çubuk grafiği oluşturulur. Alpha parametresi, çubukların şeffaflık düzeyini belirler.
**plt.bar(top_cities['name_of_city'], top_cities['population_female'], color='red', alpha=0.7, label='Female Population'):** Kırmızı renkte, kadın nüfusunu temsil eden çubuk grafiği oluşturulur.
**plt.legend():** Grafiğe bir efsane eklenir, bu efsane çubuk renklerini ve kategorilerini açıklar.

**plt.title('Top 10 Cities by Total Population'):** Grafik başlığı belirlenir.

**plt.xlabel('India Cities') ve plt.ylabel('Total Population (millions)'):** X ve Y ekseni etiketleri belirlenir.

**plt.grid(axis='y', linestyle='--', color='black', linewidth=0.5):** Y eksenindeki ızgara çizgileri belirlenir. Çizgiler kesikli (--), siyah renkte ve ince (0.5 piksel) olacaktır.

**plt.xticks(rotation=45):** X eksenindeki şehir isimlerini 45 derece döndürür.

**plt.show():** Oluşturulan grafik gösterilir.

```python
# Toplam nüfusa göre en büyük 10 şehri seçelim
top_cities = df.nlargest(10, 'population_total')

# İki bar grafiği üst üste bindirerek hem erkek hem de kadın nüfusunu gösterelim
plt.figure(figsize=(12, 6))

# Toplam nüfus bar grafiği (Yeşil)
plt.bar(top_cities['name_of_city'], top_cities['population_total'], color='green', label='Total Population')

# Erkek nüfusu (Mavi) ve Kadın nüfusu (Kırmızı) için bar grafikleri
plt.bar(top_cities['name_of_city'], top_cities['population_male'], color='blue', alpha=0.7, label='Male Population')
plt.bar(top_cities['name_of_city'], top_cities['population_female'], color='red', alpha=0.7, label='Female Population')

# Efsane (legend), başlık ve ızgara çizgileri ekleyelim
plt.legend()
plt.title('Top 10 Cities by Total Population')
plt.xlabel('India Cities')
plt.ylabel('Total Population (millions)')
plt.grid(axis='y', linestyle='--', color='black', linewidth=0.5)

# X eksenindeki etiketleri döndürelim
plt.xticks(rotation=45)

# Grafik gösterimi
plt.show()
```

```python
import numpy as np
from sklearn.metrics import pairwise_distances
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Kullanıcıdan seçilen şehir adını alalım
selected_city_name = input("Lütfen bir şehir adı girin: ")

# Seçilen şehirin veri çerçevesinde bulunup bulunmadığını kontrol edelim
if selected_city_name in df['name_of_city'].values:
    # Seçilen şehrin enlem ve boylamını bulalım
    selected_city_latitude = df.loc[df['name_of_city'] == selected_city_name, 'latitude_of_city'].values[0]
    selected_city_longitude = df.loc[df['name_of_city'] == selected_city_name, 'longitude_of_city'].values[0]

    # Tüm şehirlerin enlem ve boylam verilerini bir Numpy dizisine dönüştürelim
    city_coordinates = df[['latitude_of_city', 'longitude_of_city']].values

    # Şehirler arası uzaklıkları hesaplayalım (örneğin, Euclidean mesafesi kullanalım)
    distances = pairwise_distances(np.array([[selected_city_latitude, selected_city_longitude]]), city_coordinates, metric='euclidean')[0]

    # Şehirleri uzaklıklara göre sıralayalım ve en yakın 5 şehiri seçelim
    closest_city_indices = np.argsort(distances)[:5]

    # En yakın 5 şehrin enlem ve boylam bilgilerini alalım
    closest_city_coordinates = city_coordinates[closest_city_indices]

    # 3D nokta gösterimi için x, y ve z koordinatlarını ayarlayalım
    x = closest_city_coordinates[:, 1]  # Boylam
    y = closest_city_coordinates[:, 0]  # Enlem
    z = np.zeros(5)  # Z ekseni için sabit değer (0)

    # 3D nokta gösterimini oluşturalım
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    ax.scatter(x, y, z, c='r', marker='o')

    print(f"{selected_city_name} şehrine en yakın 5 şehir:")
    for index in closest_city_indices:
        city_name = df.loc[index, 'name_of_city']
        distance = distances[index]
        print(f"{city_name}: {distance} km")
    # Şehir adlarını işaretleyelim
    labels = [df.loc[index, 'name_of_city'] for index in closest_city_indices]
    for i, label in enumerate(labels):
        ax.text(x[i], y[i], z[i], label, fontsize=12)

    # Eksen etiketleri
    ax.set_xlabel('Boylam')
    ax.set_ylabel('Enlem')
    ax.set_zlabel('Z')

    # Grafiği gösterelim
    plt.show()

else:
    print(f"{selected_city_name} isimli bir şehir veri çerçevemizde bulunmamaktadır.")
```

```python
import numpy as np
from sklearn.metrics import pairwise_distances
import matplotlib.pyplot as plt

# Verileri tekrar işleyelim ve seçilen şehir için en yakın 5 şehiri bulalım

# Seçilen şehrin enlem ve boylamını bulalım
selected_city_latitude = df.loc[df['name_of_city'] == selected_city_name, 'latitude_of_city'].values[0]
selected_city_longitude = df.loc[df['name_of_city'] == selected_city_name, 'longitude_of_city'].values[0]

# Tüm şehirlerin enlem ve boylam verilerini bir Numpy dizisine dönüştürelim
city_coordinates = df[['latitude_of_city', 'longitude_of_city']].values

# Şehirler arası uzaklıkları hesaplayalım (Euclidean mesafesi kullanalım)
distances = pairwise_distances(np.array([[selected_city_latitude, selected_city_longitude]]), city_coordinates, metric='euclidean')[0]

# Şehirleri uzaklıklara göre sıralayalım ve en yakın 5 şehiri seçelim
closest_city_indices = np.argsort(distances)[:5]

# En yakın 5 şehrin enlem ve boylam bilgilerini alalım
closest_city_coordinates = city_coordinates[closest_city_indices]

# 2D nokta gösterimi için x ve y koordinatlarını ayarlayalım
x = closest_city_coordinates[:, 1]  # Boylam
y = closest_city_coordinates[:, 0]  # Enlem

# 2D nokta gösterimini oluşturalım
fig, ax = plt.subplots()
ax.scatter(x, y, c='r', marker='o')

print(f"{selected_city_name} şehrine en yakın 5 şehir:")
for index in closest_city_indices:
    city_name = df.loc[index, 'name_of_city']
    distance = distances[index]
    print(f"{city_name}: {distance} km")

# Şehir adlarını işaretleyelim
labels = [df.loc[index, 'name_of_city'] for index in closest_city_indices]
for i, label in enumerate(labels):
    ax.text(x[i], y[i], label, fontsize=12)

# Eksen etiketleri
ax.set_xlabel('Boylam')
ax.set_ylabel('Enlem')

# Grafiği gösterelim
plt.show()
```
