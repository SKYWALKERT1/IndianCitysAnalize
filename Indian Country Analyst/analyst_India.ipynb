### Problem:Hindistanda seçtiğimiz şehire en yakın şehirin boylam ve enlemine göre en yakın 4 tane şehiri bize getirmesi ve bunu hipotenüs (euclidean) uzaklığına göre yapması.

**Şehir Seçimi:** Kullanıcıdan bir şehir adı alınır.

**Şehir Kontrolü:** Girilen şehir adı, veri çerçevesinde (df) 'name_of_city' sütununda aranır.

**Enlem ve Boylam Bilgisi:** Seçilen şehrin enlem ve boylamı bulunur.

**Şehir Koordinatları:** Tüm şehirlerin enlem ve boylam bilgileri numpy dizisine dönüştürülür.

**Mesafe Hesaplama:** Seçilen şehir ile diğer tüm şehirler arasındaki mesafeler Euclidean mesafe metriği kullanılarak hesaplanır.

**En Yakın Beş Şehir:** Hesaplanan mesafelere göre en yakın beş şehir belirlenir.

**3D Görselleştirme:** Seçilen şehir ve en yakın beş şehir, 3D grafikte görselleştirilir. X ve Y eksenleri şehirlerin boylam ve enlem bilgilerini temsil ederken, Z ekseninde sabit bir değer (0) kullanılır.

**Sonuçların Yazdırılması:** Seçilen şehire en yakın beş şehir ve bu şehirlerle olan mesafeleri konsola yazdırılır.

**Şehir Adlarının İşaretlenmesi:** Grafik üzerinde her bir noktanın yanına ilgili şehrin adı yazılır.

**Grafiğin Gösterilmesi:** Hazırlanan 3D grafik ekranda gösterilir.

```python
import numpy as np
from sklearn.metrics import pairwise_distances
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D


selected_city_name = input("Lütfen bir şehir adı girin: ")


if selected_city_name in df['name_of_city'].values:
   
    selected_city_latitude = df.loc[df['name_of_city'] == selected_city_name, 'latitude_of_city'].values[0]
    selected_city_longitude = df.loc[df['name_of_city'] == selected_city_name, 'longitude_of_city'].values[0]

    
    city_coordinates = df[['latitude_of_city', 'longitude_of_city']].values

   
    distances = pairwise_distances(np.array([[selected_city_latitude, selected_city_longitude]]), city_coordinates, metric='euclidean')[0]

   
    closest_city_indices = np.argsort(distances)[:5]

    closest_city_coordinates = city_coordinates[closest_city_indices]

    
    x = closest_city_coordinates[:, 1]  
    y = closest_city_coordinates[:, 0]  
    z = np.zeros(5)  

    
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    ax.scatter(x, y, z, c='r', marker='o')

    print(f"{selected_city_name} şehrine en yakın 5 şehir:")
    for index in closest_city_indices:
        city_name = df.loc[index, 'name_of_city']
        distance = distances[index]
        print(f"{city_name}: {distance} km")
    
    labels = [df.loc[index, 'name_of_city'] for index in closest_city_indices]
    for i, label in enumerate(labels):
        ax.text(x[i], y[i], z[i], label, fontsize=12)

    ax.set_xlabel('Boylam')
    ax.set_ylabel('Enlem')
    ax.set_zlabel('Z')

    plt.show()

else:
    print(f"{selected_city_name} isimli bir şehir veri çerçevemizde bulunmamaktadır.")
```

```python

import numpy as np
from sklearn.metrics import pairwise_distances
import matplotlib.pyplot as plt

# Verileri tekrar işleyelim ve seçilen şehir için en yakın 5 şehiri bulalım

# Seçilen şehrin enlem ve boylamını bulalım
selected_city_latitude = df.loc[df['name_of_city'] == selected_city_name, 'latitude_of_city'].values[0]
selected_city_longitude = df.loc[df['name_of_city'] == selected_city_name, 'longitude_of_city'].values[0]

# Tüm şehirlerin enlem ve boylam verilerini bir Numpy dizisine dönüştürelim
city_coordinates = df[['latitude_of_city', 'longitude_of_city']].values

# Şehirler arası uzaklıkları hesaplayalım (Euclidean mesafesi kullanalım)
distances = pairwise_distances(np.array([[selected_city_latitude, selected_city_longitude]]), city_coordinates, metric='euclidean')[0]

# Şehirleri uzaklıklara göre sıralayalım ve en yakın 5 şehiri seçelim
closest_city_indices = np.argsort(distances)[:5]

# En yakın 5 şehrin enlem ve boylam bilgilerini alalım
closest_city_coordinates = city_coordinates[closest_city_indices]

# 2D nokta gösterimi için x ve y koordinatlarını ayarlayalım
x = closest_city_coordinates[:, 1]  # Boylam
y = closest_city_coordinates[:, 0]  # Enlem

# 2D nokta gösterimini oluşturalım
fig, ax = plt.subplots()
ax.scatter(x, y, c='r', marker='o')

print(f"{selected_city_name} şehrine en yakın 5 şehir:")
for index in closest_city_indices:
    city_name = df.loc[index, 'name_of_city']
    distance = distances[index]
    print(f"{city_name}: {distance} km")

# Şehir adlarını işaretleyelim
labels = [df.loc[index, 'name_of_city'] for index in closest_city_indices]
for i, label in enumerate(labels):
    ax.text(x[i], y[i], label, fontsize=12)

# Eksen etiketleri
ax.set_xlabel('Boylam')
ax.set_ylabel('Enlem')

# Grafiği gösterelim
plt.show()
```

**Özellik Mühendisliği (Feature Engineering):** Mevcut verilerden yeni özellikler türetilerek modelin daha kapsamlı bir veri seti üzerinde eğitilmesi sağlandı. Bu durumda, 'sex_ratio' ve 'effective_literacy_rate_total' sütunlarının çarpımından oluşan yeni bir 'interaction' özelliği eklendi. Bu yeni özellik, mevcut özellikler arasındaki etkileşimleri yakalayabilir ve modelin tahmin gücünü artırabilir.

**Veri Ölçeklendirme (Data Scaling):** Özellikler, StandardScaler kullanılarak ölçeklendirildi. Bu, özellikler arasındaki farklı değer aralıklarını normalize eder ve modelin daha hızlı ve etkili bir şekilde eğitilmesine yardımcı olur.

**Farklı Regresyon Modellerinin Denenmesi:** Ridge ve Lasso regresyonu gibi düzenlileme (regularization) tekniklerini içeren modeller kullanıldı. Bu modeller, aşırı uyuma (overfitting) karşı dirençli olabilir ve modelin genelleştirme yeteneğini artırabilir.
**MSE Değeri:MSE değeri yüksek çıktı yani bu şu anlama gelirki modelin gerçek değerlinde tahmine sapma meydana gelir bunun yanında eğitim setine yakın olmadığı için varyans'ı azdır ön yargısı çoktur varyans az olduğundan dolayı eğitim setine benzememektedir.

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Ridge, Lasso
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import StandardScaler
import pandas as pd

# Veri çerçevesini yükleyelim
df = pd.read_csv('Indian_cities_analyst.csv')

# Özellik mühendisliği: Yeni özellikler oluşturmak için mevcut verileri kullanalım
df['interaction'] = df['sex_ratio'] * df['effective_literacy_rate_total']

# Yeni özellik seti oluşturalım
X_new = df[['population_total', 'sex_ratio', 'effective_literacy_rate_total', 'interaction']]

# Özellikleri ölçeklendirelim
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_new)

# Veriyi eğitim ve test setlerine ayıralım
X_train, X_test, y_train, y_test = train_test_split(X_scaled, df['total_graduates'], test_size=0.2, random_state=42)

# Ridge Regresyon modelini deneyelim
ridge_model = Ridge(alpha=1.0)
ridge_model.fit(X_train, y_train)
ridge_pred = ridge_model.predict(X_test)
ridge_mse = mean_squared_error(y_test, ridge_pred)

# Lasso Regresyon modelini deneyelim
lasso_model = Lasso(alpha=0.1)
lasso_model.fit(X_train, y_train)
lasso_pred = lasso_model.predict(X_test)
lasso_mse = mean_squared_error(y_test, lasso_pred)

ridge_mse, lasso_mse  # Her iki modelin MSE değerlerini döndürelim
```
